```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import statsmodels.api as sm
import seaborn as sns
import scipy.stats as stats
# Imports for hierarchical clustering
from scipy.cluster.hierarchy import dendrogram, linkage  
from sklearn import preprocessing
```

```python
# Read the dataset
df = pd.read_csv("data/data3.csv")
```

```python
# We are going to use bootstrapping
sample_proportion =  0.05
dataset_size = len(df)
sample_size = sample_proportion * dataset_size

print("The sample size we are going to use will be: "+ str(sample_size))
# Run bootstrap
chosen_idx = np.random.choice(size, replace=False, size=int(sample_size))
sampled_dataset =  df.iloc[chosen_idx]
print("The new shape of the sampled dataset is: " + str(sampled_dataset.values.shape))

# Perform Linkage on the sampled dataset
arrayFormDf = sampled_dataset.values

# SkLearn Preprocess (scaling for better distance comparison)
arrayFormDf = preprocessing.scale(arrayFormDf) 

linked = linkage(arrayFormDf, 'single')

dendrogram(linked,  
            truncate_mode='lastp',
            p=12,
            leaf_rotation=45.,
            leaf_font_size=15.,
            show_contracted=True)
plt.title('Truncated Hierarchical Clustering Dendogram')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.show()  
```
